```python
#1

#Problem Statement:
#Sudhir is given an expression string exp. He has to examine whether the pairs and the orders of “{“,”}”,”(“,”)”,”[“,”]”
#are correct in exp.
#For example, he should print 'balanced' for exp = “[()]{}{[()()]()}” and 'not balanced' for exp = “[(])”

#Input:
#The first line of input contains an integer T denoting the number of test cases.
#Each test case consists of a string of expression, in a separate line.

#Output:
#Print 'balanced' without quotes if the pair of parenthesis is balanced else print 'not balanced' in a separate line.

#Example 1:
#Input
#3
#{([])}
#()
#([]
#Output
#balanced
#balanced
#not balanced



n = int(input())
for i in range(0, n):
    s = input()
    l = len(s)
    for i in range(0, l):
        if "(" in s:
            if ")" in s:
                loc = s.find("(")
                loc1 = s.find(")")
                if loc1<loc:
                    print("not balanced")
                    break
                else:
                    s = s.replace("(","")
                    s = s.replace(")","")
            else:
                print("not balanced")
                break
        elif "{" in s:
            if "}" in s:
                loc = s.find("{")
                loc1 = s.find("}")
                if loc1<loc:
                    print("not balanced")
                    break
                else:
                    s = s.replace("{","")
                    s = s.replace("}","")
            else:
                print("not balanced")
                break
        elif "[" in s:
            if "]" in s:
                loc = s.find("[")
                loc1 = s.find("]")
                if loc1<loc:
                    print("not balanced")
                    break
                else:
                    s = s.replace("[","")
                    s = s.replace("]","")
            else:
                print("not balanced")
                break
        else:
            print("balanced")
            break


# Code passed few test cases 
# But few cases still generates wrong output
```

```python
#3

#Problem Statement:
#Given an array arr[] of n integer numbers. The task is to rewrite the array by putting all multiples of 10 at the end
# of the given array.
#Note - The order of the numbers which are not multiples of 10 should remain unaltered, and similarly, the order of
#all multiples of 10 should be unaltered

#For example:
#Suppose n = 9 and arr = {10,12,5,40,30,7,50,9,10}
#You have to push all the multiples of 10 at the end of the arr[]
#Hence the output is 12 5 7 9 10 10 40 30 50 10

#Example 1 - 
#Input
#9  --value of n
#10 12 5 40 30 7 50 9 10  --elements of arr[]
#Output
#12 5 7 9 10 10 40 30 50 10



n = int(input())
l = list(map(int, input().split()))
fac = []
i = 0
while(i<len(l)):
    try:
        if l[i]%10==0:
            a = l.pop(i)
            i -= 1
            fac.append(a)
        i += 1
    except:
        pass
for i in fac:
    l.append(i)
print(l)

# Code is correct Good Job!!
```

```python
#5
#The knockout cricket tournament is about to start in a few months. All the participating teams are listed. N number of teams
#are participating in the tournament. In each game, there are 2 teams playing against each other. Only one team can be the
#winning team of the tournament. The task is to find the maximum number of games a winning team can play

#Note
#Each match is a knockout match
#2 teams can play against each other only if the difference between the matches played by them is not more than 1
#Display 'MATCH NOT POSSIBLE' if the value of n is less than 2

#Example1
#input
#3  --value of n(number of teams)
#output
#2  --this is the maximum number of matches a winning team can play

#Example2
#input
#6
#output
#3

#Example3
#input
#1
#output
#MATCH NOT POSSIBLE



# My Code
import math
n = int(input())
count = 0
if n<2:
    print('MATCH NOT POSSIBLE')
else:
    while n!=1:
        if n%2 == 0:
            n = n/2
            count += 1
        else:
            n = math.ceil(n/2)
            count += 1
    print(count)
# Code is correct Good Job!!!
```

```python
# 6

# Encoding data
# 2 line of input is given, 1st has to be encode and 2nd has to decode
# Encoding - Convert the characters of string into ascii code, add 5 in it and again convert it back to character
# Decoding - Convert the characters of string into ascii code, subtract 5 in it and again convert it back to character

# If string contains any number or special character print INVALID INPUT

# Every word in input is seperated by ", " and decoded msg generated should also be seperated by it. Same goes while decoding

# Example 
# Input
# Hello, Sir    --(s1)
# Mjqqt, Rjjy   --(s2)

# Output
# Mjqqt, Xnw    --(Encoded version of s1)
# Hello, Meet   --(Decoded version of s2)


s_encoding = input()
encode = s_encoding.split(", ")
encoded_string = []
for s in encode:
    code = []
    answer = []
    if s.isalpha():
        for i in s:
            c = ord(i)
            code.append(c+5)
        for i in code:
            c = chr(i)
            answer.append(c)
        ans = ''
        ans = ans.join(answer)
        encoded_string.append(ans)
    else:
        print('INVALID INPUT')
ans = ', '
ans = ans.join(encoded_string)
print(ans)

s_decoding = input()
decode = s_decoding.split(", ")
decoded_string = []
for s in decode:
    code = []
    answer = []
    if s.isalpha():
        for i in s:
            c = ord(i)
            code.append(c-5)
        for i in code:
            c = chr(i)
            answer.append(c)
        ans = ''
        ans = ans.join(answer)
        decoded_string.append(ans)
    else:
        print('INVALID INPUT')
ans = ', '
ans = ans.join(decoded_string)
print(ans)

# Code is correct Good Job!!!
```

```python
#10

#Let us find out whether the sum of the digits of the given positive integer number N is UNO or not. 
#Given a positive integer number N, reduce the number of digits of N by computing the sum of all the digits to get a new number.
#If this new number exceeds 9, then sum the digits of this new number to get another number and continue this way until a 
#single digit value is obtained as the ‘digit sum’. The task here is to find out whether the result of the digit sum done 
#this way is ‘1’ or not.

# Example
# Input - 
# 51112

# Output - 
# UNO

# Explaination - 
# 5+1+1+1+2 = 10, 10 is more than 9, therefore
# 1+0 = 1, Sum is 1 therefore print 'UNO'


def digit_sum(number):
    str_num = str(number)
    total = 0
    for i in str_num:
        total += int(i)
    return total

n = int(input())
ans = digit_sum(n)

while ans>9:
    ans = digit_sum(ans)

if ans == 1:
    print("UNO")
else:
    print("NOT UNO")

# Code is correct Good Job!!!
```

```python
# 17

# input a list of elements..
# Task is to output the maximum index product. 
# For every index 'j' , index product = left(j)*right(j)
# left(j) = and index 'k' which is closest to index 'j', such that k<j and arr[k]>arr[j], if not such index exist then left(j)=0
# right(j) = and index 'I' which is closest to index 'j', such that I>j and arr[I]>arr[j], if not such index exist then right(j)=0

# Example
# Input -
# -3 4 3 6 4 5 -2

# Output
# 15

# Explanation - 
# Index Product of index 0 = 0*1 = 0
# Index Product of index 1 = 0*3 = 0
# Index Product of index 2 = 1*3 = 3
# Index Product of index 3 = 0*0 = 0
# Index Product of index 4 = 3*5 = 15
# Index Product of index 5 = 3*0 = 0
# Index Product of index 6 = 5*0 = 0
# Thus max. product is 15

# Example 
# Input - 
# 1 5 4 3 5 2
# Output - 
# 8



elements = list(map(int, input().split()))
ans = []

for i in range(0, len(elements)):
    if i == 0 or i == len(elements)-1:
        ans.append(0)
    else:
        left = 0
        right = 0
        for j in range(i+1, len(elements)):
            if elements[j]>elements[i]:
                right = j
                break
            else:
                pass
        for j in range(i-1, 0, -1):
            if elements[j]>elements[i]:
                left = j
                break
            else:
                pass
        ans.append(left*right)

print(max(ans))

# Code is correct Good Job!!!
```

```python
# 19

# A control device has 4 buttons that can be used to move a character around a screen in 4 directions: Up (U), Down (D), 
# Left (L), and Right (R). The movement needs to be optimized by deleting unnecessary instructions while maintaining the same 
# destination. Given the original set of instructions, what is the reach the destination?

# Example 
# input - URDR
# output - 2

# explanation - Given an original set of instructions s 'URDR', the final destination is 2
# units to the right of the initial position after the character moves up, right, down, and right. If 'U and D are deleted,
# the destination remains the same. The answer 2 will be returned

# Example 
# input - RRR
# output - 0

# explanation - there is nothing that can be deleted from the string, thus the output is 0.



s = input()
answer = 0

u = s.count('U')
d = s.count('D')
r = s.count('R')
l = s.count('L')

if u>d:
    answer += d*2
else:
    answer += u*2

if r>l:
    answer += l*2
else:
    answer += r*2

print(answer)

# Code is correct Good Job!!!
```
